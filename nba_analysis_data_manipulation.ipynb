```python
import pandas as pd
import numpy as np
player_stats = pd.read_csv('players_stats_by_season.csv')
players = pd.read_csv('players.csv')
salaries = pd.read_csv('salaries.csv')
```

```python
player_stats = player_stats[(player_stats['League'] == 'NBA') & (player_stats['Stage'] == 'Playoffs')]
player_stats['player_key'] = player_stats['Player'] + ' (' + player_stats['Season'] + ')'
player_stats
```

```python
player_keys = {}
for i in range(0, len(players)):
    row = players.iloc[i]
    player_keys[row['_id']] = row['name']
```

```python
full_names = []
season_mod = []
for i in range(0, len(salaries)):
    row = salaries.iloc[i]

    season_mod.append(str(row['season_start']) + ' - ' + str(row['season_end']))
    full_names.append(player_keys[row['player_id']])
salaries['name'] = full_names
```

```python
salaries['season'] = season_mod
```

```python
salaries
```

```python
salaries['player_key'] = salaries['name'] + ' (' + salaries['season'] + ')'
```

```python
salaries
```

```python
new_df = player_stats.merge(salaries, left_on='player_key', right_on='player_key')
new_df
```

```python
new_df.columns
```

```python
final_df = new_df.drop(columns=['birth_year', 'birth_month', 'birth_date', 'height', 'height_cm', 'weight', 'weight_kg', 'nationality', 'high_school', 'draft_round', 'draft_pick', 'draft_team', 'player_id', 'League', 'Season'])
```

```python
final_df
```

```python
szn_stats = pd.read_csv('Seasons_Stats.csv')
szn_stats
```

```python
positions = {}
for i in range(0, len(szn_stats)):
    row = szn_stats.iloc[i]
    positions[row['Player']] = row['Pos']
positions["Shaquille O'Neal"] = 'C'
```

```python
positions["Reggie Miller"] = 'SG'
positions['Karl Malone'] = 'SG'
positions['Allen Iverson'] = 'PG'
positions['Dikembe Mutombo'] = 'C'
positions['Metta World Peace'] = 'SF'
positions['Yao Ming'] = 'C'
positions['Scottie Pippen'] = 'SF'
positions['John Stockton'] = 'PG'
positions['Bam Adebayo'] = 'C'
positions['Jayson Tatum'] = 'PF'
positions['Donovan Mitchell'] = 'PG'
```

```python
pos = []
for i in range(0, len(final_df)):
    row = final_df.iloc[i]
    if row['name'] in positions: 
        pos.append(positions[row['name']])
    else: 
        pos.append('None')
final_df['position'] = pos
```

```python
final_df.columns
```

```python
final_df['points_per_game'] = final_df['PTS'] / final_df['GP']
final_df['rebounds_per_game'] = final_df['REB'] / final_df['GP']
final_df['assists_per_game'] = final_df['AST'] / final_df['GP']
final_df['blocks_per_game'] = final_df['BLK'] / final_df['GP']
final_df['steals_per_game'] = final_df['STL'] / final_df['GP']
final_df['turnovers_per_game'] = final_df['TOV'] / final_df['GP']
```

```python
final_df['player_utility'] = final_df['points_per_game'] + 1.2 * final_df['rebounds_per_game'] + 1.5 * final_df['assists_per_game'] + 2 * final_df['blocks_per_game'] + 2 * final_df['steals_per_game'] - final_df['turnovers_per_game']
```

```python
final_df.columns
```

```python
final_df = final_df.drop(columns=['GP', 'MIN', 'FGM', 'FGA', '3PM', '3PA', 'FTM', 'TOV', 'PF', 'ORB', 'DRB', 'REB', 'AST', 'STL', 'BLK', 'PTS', 'FTA', 'Player'])
```

```python
final_df = final_df.rename(columns={"Stage": "stage", "Team": "team"})
```

```python
final_df = final_df.drop(columns=['Player'])
```

```python
final_df
```

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv')  

final_df.to_csv('out.zip', index=False,
          compression=compression_opts)  
```
